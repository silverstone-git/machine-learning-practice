```python
import torch
from transformers import BitsAndBytesConfig, TrainingArguments, AutoTokenizer, AutoModelForCausalLM
from peft import LoraConfig, get_peft_model, prepare_model_for_kbit_training, PeftModel
from trl import SFTTrainer
from datasets import Dataset
```

```python
with open("train.txt", "r", encoding="utf-8") as f:
    peculiar_text = f.read()

# hf dataset making
data = [{"text": peculiar_text}]
dataset = Dataset.from_list(data)
```

```python
dataset
```

```python
!huggingface-cli login
```

```python
# 1.7B, and 4B model isnt fitting in the 14.7GB VRAM of the free version, even in r= 8
# model_id = "Qwen/Qwen3-4B"
# model_id= "Qwen/Qwen3-1.7B"
# model_id= "Qwen/Qwen3-0.6B" -> this works for r64
model_id= "google/gemma-3-1b-pt"
tokenizer = AutoTokenizer.from_pretrained(model_id)

if tokenizer.pad_token is None:
    tokenizer.pad_token = tokenizer.eos_token


model = AutoModelForCausalLM.from_pretrained(
    model_id,
    device_map="auto", # Automatically maps model to available devices (GPU/CPU)
    torch_dtype= torch.bfloat16,
    trust_remote_code=True, # Set to True if the model requires custom code (check model card)
    attn_implementation= 'eager', # SEt this if using gemma
)
```

```python
# Quantization Configuration (for QLoRA)
bnb_config = BitsAndBytesConfig(
    load_in_4bit=True,
    bnb_4bit_quant_type="nf4",        # Quantization type: NormalFloat 4-bit
    bnb_4bit_use_double_quant=True,   # Double quantization for memory efficiency
    bnb_4bit_compute_dtype=torch.bfloat16, # Compute dtype for faster training
)


# Prepare model for k-bit training (important for QLoRA)
model = prepare_model_for_kbit_training(model)
```

```python

lora_config = LoraConfig(
    r=64, # LoRA rank. Common values are 8, 16, 32, 64. Higher means more trainable parameters, potentially better performance but more VRAM.
    lora_alpha=128, # LoRA scaling factor. Usually 2 * r.
    target_modules=[ # Which modules of the model to apply LoRA to.
        "q_proj",
        "k_proj",
        "v_proj",
        "o_proj",
        "gate_proj",
        "up_proj",
        "down_proj",
        "lm_head", # to fine-tune the output layer, often helpful for style.
    ],
    bias="none", # "none", "all", or "lora_only". "none" is common for memory.
    lora_dropout=0.05, # Dropout for LoRA layers to prevent overfitting.
    task_type="CAUSAL_LM", # Specifies the task type (Causal Language Modeling for text generation)
)


model = get_peft_model(model, lora_config)
model.print_trainable_parameters() # This will show how few parameters are actually trained.



output_dir = "./dickbot"
training_args = TrainingArguments(
    output_dir=output_dir,
    per_device_train_batch_size=4, # Adjust based on your GPU VRAM. Start small.
    gradient_accumulation_steps=2, # Accumulate gradients over multiple steps to simulate larger batch size.
    warmup_steps=50, # Number of steps for learning rate warm-up.
    max_steps=250, # Total number of training steps. Adjust based on dataset size and desired convergence.
    learning_rate=1e-4, # Optimal learning rate for LoRA is often higher than full fine-tuning.
    fp16=False, # Set to True if your GPU supports it and bnb_4bit_compute_dtype is torch.float16, or bfloat16=True.
    bf16=True, # Set to True if your GPU supports bfloat16 (e.g., Ampere architecture or newer). Highly recommended.
    logging_steps=10, # Log training progress every N steps.
    optim="paged_adamw_8bit", # Optimizer, often good for QLoRA.
    #optim= "adamw_torch",
    weight_decay=0.01,
    lr_scheduler_type="cosine", # Learning rate scheduler.
    save_strategy="steps",
    save_steps=100, # Save checkpoint every N steps.
    push_to_hub= False,# Set to True if you want to push to Hugging Face Hub after training.
    report_to="tensorboard", # Report metrics to TensorBoard for visualization.
    # evaluation_strategy="steps", # Uncomment if you have an eval_dataset
    # eval_steps=100,
)

######## TOOK JUST 170 TRAINING STEPS TO GET TO 0 LOSS (r= 16, 0.6B params)

# Initialize SFTTrainer
# For pure style transfer where the model just needs to learn the "peculiar_text",
# you can use dataset_text_field='text'. If you had instruction/response pairs,
# you'd define a formatting function for 'packing=True'.

trainer = SFTTrainer(
    model=model,
    train_dataset=dataset,
    # eval_dataset=eval_dataset, # Uncomment if you have an eval_dataset
    # dataset_text_field="text", # The column in your dataset containing the text.
    #max_seq_length= 1024, # Maximum sequence length for training. Adjust based on your essays' length and VRAM.
    args=training_args,
)
```

```python
print("Starting training...")
trainer.train()
print("Training complete!")
```

```python
# Saving the Fine-tuned Adapter Weights
# The LoRA weights are saved separately from the base model.
trainer.model.save_pretrained(output_dir)
tokenizer.save_pretrained(output_dir)

print(f"LoRA adapters saved to {output_dir}")

# Merging LoRA Adapters with the Base Model for Easier Inference
# This creates a full model with the fine-tuned weights merged in.
# It requires enough VRAM to load the original base model + the merged weights.
# You might need to load the base model again in full precision for this.
del model # Clear the quantized model from memory
torch.cuda.empty_cache() # Clear CUDA cache

# Loading the base model in full precision (or bfloat16 if supported)
base_model = AutoModelForCausalLM.from_pretrained(
    model_id,
    device_map="auto",
    torch_dtype=torch.bfloat16,
    trust_remote_code=True,
)

# Load the LoRA adapters
lora_model = PeftModel.from_pretrained(base_model, output_dir)

# Merging and unloading (this creates a new model with merged weights)
merged_model = lora_model.merge_and_unload()
print("LoRA adapters merged with base model.")

# Save the merged model (full model) and tokenizer
merged_model_save_path = f"{output_dir}_merged"
merged_model.save_pretrained(merged_model_save_path)
tokenizer.save_pretrained(merged_model_save_path)
print(f"Merged model saved to {merged_model_save_path}")
```

```python
# Inference (Testing the Fine-tuned Model)
from transformers import pipeline

# You can load either the `merged_model` or the `base_model` + `lora_adapter`
# For convenience, let's use the merged model if you saved it.
# If you didn't merge, you'd load the base model, then load the adapter weights on top.

print("\n--- Testing Fine-tuned Model ---")
# Use the merged model if available, otherwise the PEFT model directly
inference_model = merged_model if 'merged_model' in locals() else trainer.model

pipe = pipeline(
    "text-generation",
    model=inference_model,
    tokenizer=tokenizer,
    torch_dtype=torch.float32,
)

# Test prompts
test_prompts = [
    "Write a short paragraph about the fleeting nature of time.",
    "Describe a common object, like a chair.",
    "Begin a narrative about a journey into an unknown land of curiosity.",
]

for prompt_text in test_prompts:
    print(f"\nPrompt: {prompt_text}")
    # Adjust generation parameters to encourage longer, more stylistic output
    outputs = pipe(
        prompt_text,
        max_new_tokens=128,
        num_return_sequences=1,
        do_sample=True,
        temperature=0.8, # Higher temperature for more creativity
        top_k=50,
        top_p=0.95,
        repetition_penalty=1.2, # Helps prevent repetitive phrasing
        eos_token_id=tokenizer.eos_token_id
    )
    generated_text = outputs[0]["generated_text"]
    # Remove the prompt from the generated text if it's prepended
    if generated_text.startswith(prompt_text):
        generated_text = generated_text[len(prompt_text):].strip()
    print(f"Generated: {generated_text}")
```

```python
# Inference (Testing the Fine-tuned Model)
from transformers import pipeline

# You can load either the `merged_model` or the `base_model` + `lora_adapter`
# For convenience, let's use the merged model if you saved it.
# If you didn't merge, you'd load the base model, then load the adapter weights on top.

print("\n--- Testing Fine-tuned Model ---")
# Use the merged model if available, otherwise the PEFT model directly
inference_model = merged_model if 'merged_model' in locals() else trainer.model

pipe = pipeline(
    "text-generation",
    model=inference_model,
    tokenizer=tokenizer,
    torch_dtype=torch.float32,
)

# Test prompts
test_prompts = [
    "Write a short paragraph about the fleeting nature of time.",
    "Describe a common object, like a chair.",
    "Begin a narrative about a journey into an unknown land of curiosity.",
]

for prompt_text in test_prompts:
    print(f"\nPrompt: {prompt_text}")
    # Adjust generation parameters to encourage longer, more stylistic output
    outputs = pipe(
        prompt_text,
        max_new_tokens=128,
        num_return_sequences=1,
        do_sample=True,
        temperature=0.8, # Higher temperature for more creativity
        top_k=50,
        top_p=0.95,
        repetition_penalty=1.2, # Helps prevent repetitive phrasing
        eos_token_id=tokenizer.eos_token_id
    )
    generated_text = outputs[0]["generated_text"]
    # Remove the prompt from the generated text if it's prepended
    if generated_text.startswith(prompt_text):
        generated_text = generated_text[len(prompt_text):].strip()
    print(f"Generated: {generated_text}")
```
